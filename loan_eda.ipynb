```python
Q1 = final_df["loan_amount"].quantile(0.25)
Q3 = final_df["loan_amount"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = final_df[
    (final_df["loan_amount"] < lower) |
    (final_df["loan_amount"] > upper)
]

print("Total Outliers:", len(outliers))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert submission_date to datetime
final_df['submission_date'] = pd.to_datetime(final_df['submission_date'])

# Create Month column
final_df['Month'] = final_df['submission_date'].dt.to_period('M').astype(str)

# Monthly aggregation
monthly = final_df.groupby('Month').agg({
    'application_id':'count',
    'Processing_Days':'mean'
}).reset_index()

monthly.rename(columns={
    'application_id':'Loan_Volume'
}, inplace=True)

monthly
```

```python
plt.figure(figsize=(12,5))

plt.plot(monthly['Month'], monthly['Loan_Volume'], label='Loan Volume')

plt.plot(monthly['Month'], monthly['Processing_Days'], label='Avg Processing Days')

plt.xticks(rotation=45)

plt.legend()

plt.title("Loan Volume vs Processing Days")

plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
monthly_ts = monthly.copy()

monthly_ts['Month'] = pd.to_datetime(monthly_ts['Month'])

monthly_ts.set_index('Month', inplace=True)

result = seasonal_decompose(
    monthly_ts['Loan_Volume'],
    model='additive',
    period=12
)

result.plot();
```

```python
from sklearn.preprocessing import LabelEncoder

loan_encoder = LabelEncoder()
final_df["LoanType_Encoded"] = loan_encoder.fit_transform(final_df["loan_type_y"])
```

```python
X = final_df[
    [
        "loan_amount",
        "credit_score",
        "interest_rate",
        "experience_years",
        "Region_Encoded",
        "Officer_Encoded",
        "LoanType_Encoded"
    ]
]

y = final_df["Processing_Days"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

grid_search = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_grid=param_grid,
    cv=3,
    scoring='r2',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
```

```python
best_rf = grid_search.best_estimator_

predictions = best_rf.predict(X_test)

from sklearn.metrics import mean_absolute_error, r2_score

mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("Optimized MAE:", round(mae,2))
print("Optimized R²:", round(r2,4))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': best_rf.feature_importances_
})

importance = importance.sort_values(
    by='Importance',
    ascending=False
)

plt.figure(figsize=(8,5))
plt.bar(importance['Feature'], importance['Importance'])
plt.title("Feature Importance")
plt.xlabel("Features")
plt.ylabel("Importance")
plt.xticks(rotation=45)
plt.show()
```
